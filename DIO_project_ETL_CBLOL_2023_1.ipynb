#Extract

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("teams_stats_cblol_2023_1.csv")

teams = data["Name"]
win_rate = data["Win rate"]
k_d = data["K:D"]

print("Teams:", teams)
print("Win rate:", win_rate)
print("K:D:", k_d)
```

#Tranform



```python
team_rank = data.groupby('Name')[['K:D', 'Win rate']].mean().reset_index()

team_rank = team_rank.sort_values(by='K:D', ascending=False)

print(team_rank)
```

#Load

```python
plt.figure(figsize=(12, 6))

sns.barplot(x='K:D', y='Name', data=team_rank, color='red')

plt.xlabel('Média do time de K:D')
plt.ylabel('Time')
plt.title('Desempenho Médio por K:D por Time')
plt.tight_layout()
```
